## Moving and Weighted Moving Average 

```python
#prints the mean absolute error for moving average and weighted moving average 
df2 = df[['day2', 'day3', 'day4', 'day5', 'day6', 'day7', 'day8', 'day9', 'day10', 'day11', 'day12', 'day13', 'day14']]
df5 = df[['day2', 'day3', 'day4', 'day5']]
slist2 = []
slist5 = []
for i in range(0, df.shape[0]):
    slist2.append(df2[i:i+1].sum().sum()/13)
    slist5.append(df5[i:i+1].sum().sum()/4)
print("Mean absolute error: %.2f" % mean_absolute_error(slist2, df.day1))
print("Mean absolute error: %.2f" % mean_absolute_error(slist5, df.day1))
```

## ARIMA

```python
#prints the mean absolute error for ARIMA
predlist = []
for i in range(0, df2.shape[0]):
    mod = ARIMA(list(df2.iloc[i]), order=(0, 0, 0)).fit()
    fc, se, conf = mod.forecast(3, alpha=0.05)
    predlist.append(fc)
    
print("Mean absolute error: %.2f" % mean_absolute_error(predlist, df.day1))
```
