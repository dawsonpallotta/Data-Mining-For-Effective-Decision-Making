```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.utils import resample
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn import tree
import xgboost as xgb
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#import and read data csv
df1 = pd.read_csv("C:/Users/student/Data Minning/featuresDefault.csv") 
print(df1.shape)
df1.head(2)
```

```python
#creates models using cross-validation and scored with MAE
folds = 3

reg = LinearRegression()
lasso = Lasso()
ridge = Ridge() 

scoresreg = cross_val_score(reg, df1[df1.columns[3:]],  df1[df1.columns[1]], cv=folds, scoring = 'neg_mean_absolute_error')
scoreslasso = cross_val_score(lasso, df1[df1.columns[3:]],  df1[df1.columns[1]], cv=folds, scoring = 'neg_mean_absolute_error')
scoresridge = cross_val_score(ridge, df1[df1.columns[3:]],  df1[df1.columns[1]], cv=folds, scoring = 'neg_mean_absolute_error')

print(scoresreg)
print(scoresreg.mean())
print(scoreslasso)
print(scoreslasso.mean())
print(scoresridge)
print(scoresridge.mean())
```

```python
#creates models using cross-validation and scored with RMSE
folds = 3

reg = LinearRegression()
lasso = Lasso()
ridge = Ridge() 

scoresreg = cross_val_score(reg, df1[df1.columns[3:]],  df1[df1.columns[1]], cv=folds, scoring = 'neg_root_mean_squared_error')
scoreslasso = cross_val_score(lasso, df1[df1.columns[3:]],  df1[df1.columns[1]], cv=folds, scoring = 'neg_root_mean_squared_error')
scoresridge = cross_val_score(ridge, df1[df1.columns[3:]],  df1[df1.columns[1]], cv=folds, scoring = 'neg_root_mean_squared_error')

print(scoresreg)
print(scoresreg.mean())
print(scoreslasso)
print(scoreslasso.mean())
print(scoresridge)
print(scoresridge.mean())
```

```python
#linear regression using holdout
df_train, df_test = train_test_split(df1, test_size=0.33, random_state = 33)
reg.fit(df_train[df_train.columns[3:]], df_train[df_train.columns[1]])
regpred = reg.predict(df_test[df_test.columns[3:]])
print("Coefficients: \n", reg.coef_)
print("Mean squared error: %.2f" % mean_squared_error(df_test[df_test.columns[1]], regpred))
print("Coefficient of determination: %.2f" % r2_score(df_test[df_test.columns[1]], regpred))
```

```python
#holdout using lasso regression
df_train, df_test = train_test_split(df1, test_size=0.33, random_state = 33)
lasso.fit(df_train[df_train.columns[3:]], df_train[df_train.columns[1]])
lassopred = lasso.predict(df_test[df_test.columns[3:]])
print("Coefficients: \n", lasso.coef_)
print("Mean squared error: %.2f" % mean_squared_error(df_test[df_test.columns[1]], lassopred))
print("Coefficient of determination: %.2f" % r2_score(df_test[df_test.columns[1]], lassopred))
```

```python
#holdout using ridge regression
df_train, df_test = train_test_split(df1, test_size=0.33, random_state = 33)
ridge.fit(df_train[df_train.columns[3:]], df_train[df_train.columns[1]])
ridgepred = ridge.predict(df_test[df_test.columns[3:]])
print("Coefficients: \n", ridge.coef_)
print("Mean squared error: %.2f" % mean_squared_error(df_test[df_test.columns[1]], ridgepred))
print("Coefficient of determination: %.2f" % r2_score(df_test[df_test.columns[1]], ridgepred))
```
